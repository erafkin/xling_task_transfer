```python
import requests

langs = ["eng", "deu", "fra", "spa", "hin"]
base = "https://raw.githubusercontent.com/unimorph/"

def load_lang(lang):
    url = f"{base}{lang}/master/{lang}"
    text = requests.get(url).text.strip().split("\n")
    return [line.split("\t") for line in text]

data = {lang: load_lang(lang) for lang in langs}

print(data["eng"][:5])
```
