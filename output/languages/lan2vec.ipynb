```python
# NOTE: in order to run this you need to pull the distances zip file and drop it into the lang2vec installation as described in https://github.com/antonisa/lang2vec
# or potentially just unzip the distances.zip file if that gets installed when you install directly from the repository
import lang2vec.lang2vec as l2v
distances = {}
types = ["genetic", "geographic", "syntactic", "inventory", "phonological", "featural"]
for t in types:

    distances[t] = {}
    distances[t]["es"] = l2v.distance(t, "eng", "spa")
    distances[t]["de"] = l2v.distance(t, "eng", "deu")
    distances[t]["hi"] = l2v.distance(t, "eng", "hin")
    distances[t]["zh"] = l2v.distance(t, "eng", "zho")

distances
```

```python
import json
with open("./output/languages/lang2vec.json", "w") as f:
    json.dump(distances, f, indent=4)
```
