```python
from urielplus import urielplus
import json

u = urielplus.URIELPlus()
distances = {}
types = ["genetic", "geographic", "syntactic", "phonological", "featural"]
for t in types:

    distances[t] = {}
    distances[t]["es"] = u.new_distance(t, ["eng", "spa"])
    distances[t]["de"] = u.new_distance(t, ["eng", "deu"])
    distances[t]["hi"] = u.new_distance(t, ["eng", "hin"])
    distances[t]["zh"] = u.new_distance(t, ["eng", "zho"])
    distances[t]["fr"] = u.new_distance(t, ["eng", "fra"])
    distances[t]["ru"] = u.new_distance(t, ["eng", "rus"])

with open("uriel_plus.json", "w") as f:
    json.dump(distances, f, indent=4)
```
